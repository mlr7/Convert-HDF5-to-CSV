```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import os

import h5py

import zipfile
```

#### Unzip raw data files (bunch of hd5f data files)

```python
archive = zipfile.ZipFile('./DATASETchallenge.zip')

for file in archive.namelist():
    #if file.startswith('foo/'):
    #    archive.extract(file, 'destination_path')
    archive.extract(file, './data_DATASETchallenge/')
```

#### Convert from hdf5 to csv

```python
ls_data_hdf5 = os.listdir('./DATASETchallenge')
print(len(ls_data_hdf5))

ls_data_hdf5[0:5] # start count from 1 (not 0) to avoid 'labels.txt'
```

```python
df_all = []

df_all = pd.DataFrame(columns= ['x', 'y', 'NominalPower', 'NominalSpeed',
                  'NominalSpotDiameter', 'LaserPowerCurrent',
                  'SignalPdInGaAs', 'IDbulkLayer', 'IDoocLayer',
                  'c_1_cost', 'c_2_cost'
                 ])

for i in range(1,len(ls_data_hdf5)):
    print(i)
    #print(ls_data_hdf5[i])
    sample_dataset_object = h5py.File('./DATASETchallenge/' + ls_data_hdf5[i], 'r')
    list(sample_dataset_object.keys())
    sample_dataset = sample_dataset_object['OpenData'][:]
    sample_dataset = sample_dataset.transpose()
    # print(sample_dataset.shape)df = []

    df = pd.DataFrame(sample_dataset)

    df.columns = ['x', 'y', 'NominalPower', 'NominalSpeed',
                  'NominalSpotDiameter', 'LaserPowerCurrent',
                  'SignalPdInGaAs', 'IDbulkLayer', 'IDoocLayer',
                  'c_1_cost', 'c_2_cost'
                 ]
    #print(ls_data_hdf5[i])
    df['file'] = ls_data_hdf5[i]
    #print(df_all.shape)
    df_all = pd.concat([df_all, df])
    print(df_all.shape)
```

```python
df_all = df_all.reset_index(drop=True)
```

```python
display(df_all.head())
display(df_all.tail())
```

### Step 2: Add in the sequence numbering

```python
df = df_all.copy()
print(df.shape)
```

```python
# create layer feature
df['layer'] = df['file'].str.replace('layer', '')
df['layer'] = df['layer'].str.replace('.hdf5', '').astype(int)
```

```python
display(df.head())
display(df.tail())
```

```python
ls_uniq = df['layer'].unique()
ls_uniq
```

```python
ls_uniq.sort()
# ls_uniq[0:10]
ls_uniq
```

```python
df2 = []
df2 = pd.DataFrame(columns= ['x', 'y', 'NominalPower', 'NominalSpeed',
                  'NominalSpotDiameter', 'LaserPowerCurrent',
                  'SignalPdInGaAs', 'IDbulkLayer', 'IDoocLayer',
                  'c_1_cost', 'c_2_cost', 'file', 'layer', 'step'
                 ])
df2
```

```python
for layer in ls_uniq: #[0:10]:
    print(layer)
    df_tmp = []
    df_tmp = df[df['layer'] == layer]
    df_tmp = df_tmp.reset_index(drop=True)
    df_tmp['step'] = df_tmp.index.to_list()
    df_tmp['step'] = df_tmp['step'] + 1
    df2 = pd.concat([df2, df_tmp])

df2 = df2.reset_index(drop=True)
```

```python
display(df2.head())
display(df2.tail())
```

```python
df2.to_csv('data_additivemfg_LPBF_convertedtrumpfchallenge_v20230403a.csv', index=False)
```
